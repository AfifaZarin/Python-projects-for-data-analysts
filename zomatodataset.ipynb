```python
final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'RatingCount'})
```

```python
ratings
```

## Observation: 
1. when rating is between 4.5 to 4.9---> Excellent
2. when the rating are between 4.0 to 4.4---> very good
3. when the rating is between 3.5 to 3.9--->good
4. ehn the rating is between 2.5 to 3.4--->average
4. ehn the rating is between 1.8 to 2.4--->poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='RatingCount',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='RatingCount',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1. no rated count is very high
2. maximum no of rating are between 2.5 to 3.4

```python
##countplot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```
